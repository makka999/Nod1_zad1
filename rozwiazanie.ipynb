```python
# 28. Grupowanie wierszy według wartości kolumny kategoryzowanej, potem uśrednienie wartości
grouped = df.groupby('gender').mean(numeric_only=True)
print(grouped)
```

```python
# 29. Grupowanie wierszy według wartości kolumny kategoryzowanej, potem
# - uśrednienie wartości dla pewnych kolumn, liczba wartości i mediana dla pozostałych
grouped_agg = ramka1.groupby('Name').agg({'Age': ['mean', 'count'], 'Name': ['max']})
print(grouped_agg)
```

```python
# 30. Wyświetl nazwy kolumn indeksu złożonego
print(grouped_agg.columns)
```

```python
# 31. Sortowanie kolumn indeksu złożonego
grouped_agg_sorted = grouped_agg.sort_index(axis=1)
print(grouped_agg_sorted)
```

```python
# 32. Tworzenie tabeli przestawnej (pivot table)
pivot = ramka1.pivot_table(values='Age', index='Surname', aggfunc='mean')
print(pivot)
```

```python
# 33. Wyświetlenie indeksów i kolumn tabeli przestawnej
print("Indeksy:", pivot.index)
print("Kolumny:", pivot.columns)
```

```python
# 34. Utwórz indeks złożony tabeli przestawnej i wyświetl go
pivot_multi = ramka1.pivot_table(values='Age', index=['Surname', 'Name'], aggfunc='mean')
print(pivot_multi)
```
